### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data["protocol_type"] = labelencoder.fit_transform(data["protocol_type"])
data["service"] = labelencoder.fit_transform(data["service"])
data["flag"] = labelencoder.fit_transform(data["flag"])
data["class"] = labelencoder.fit_transform(data["class"])
```

# One-Hot-Encoding

```python
data=pd.concat([data,pd.get_dummies(data["protocol_type"],prefix="protocol_type")],axis=1).drop(["protocol_type"],axis=1)
data=pd.concat([data,pd.get_dummies(data["service"],prefix="service")],axis=1).drop(["service"],axis=1)
data=pd.concat([data,pd.get_dummies(data["flag"],prefix="flag")],axis=1).drop(["flag"],axis=1)
```

```python
import h2o
import numpy as np
import math
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.grid.grid_search import H2OGridSearch
h2o.init(nthreads=-1, strict_version_check=True)
## optional: connect to a running H2O cluster
#h2o.init(ip="mycluster", port=55555)
```

```python
## 'path' can point to a local file, hdfs, s3, nfs, Hive, directories, etc.
df = h2o.import_file(path = "C:\\Users\\AJIT\\Desktop\\Data science\\RP\\NSL-KDD\\Network-Intrusion-Detection-master\\df_train.csv")
df_test = h2o.import_file(path ="C:\\Users\\AJIT\\Desktop\\Data science\\RP\\NSL-KDD\\Network-Intrusion-Detection-master\\df_test.csv")
```

```python
response = "class"
df[response] = df[response].asfactor()           
df_test[response] = df_test[response].asfactor()           
predictors = df.columns
predictors[1:123]
```
